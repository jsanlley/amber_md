```python
#Define active site and interface inhibitors

#Combine all interface residues into one string
interface_a = [1,4,10,12,14,139,290,299,166]
interface_b = [x+306 for x in interface_a]
interface_all = []
interface_all.extend(interface_a)
interface_all.extend(interface_b)

interface_sel = ''

for x in interface_all:
    interface_sel += str(x) + ' '
    
active_a = [41,49,140,145,189,163,166,172]
active_b = [x+306 for x in active_a]

def make_atom_string(selection_list):
    selection = ''
    for x in selection_list:
        selection += str(x) + ' '
        
    return selection
```

# Compare POVME sphere size (10 vs. 15) and distribution plots resolution (visualized)

```python
def plot_distribution(apo_dimer,nat_dimer,nirm_dimer,ens_dimer,title):
    
    params = {
        'legend.fontsize': '12',
        'axes.labelsize': '18',
         'axes.titlesize': '24',
         'xtick.labelsize': '18',
         'ytick.labelsize': '18'
    }

    pylab.rcParams.update(params)
    
    #Set range
    collection = apo_dimer + nat_dimer + nirm_dimer + ens_dimer
    start = np.min(collection) - 1
    stop = np.max(collection) + 1
    
    domain = np.linspace(start,stop,10000)
    
    #Calculate distribution
    dens_apo = gaussian_kde(apo_dimer)
    dens_apo.covariance_factor = lambda : 0.25
    
    dens_nat = gaussian_kde(nat_dimer)
    dens_nat.covariance_factor = lambda : 0.25
    
    dens_nirm = gaussian_kde(nirm_dimer)
    dens_nirm.covariance_factor = lambda : 0.25
    
    dens_ens = gaussian_kde(ens_dimer)
    dens_ens.covariance_factor = lambda : 0.25
    
    #Plot
    fig, axs = plt.subplots(1,1, sharex=True, sharey=True, tight_layout=True, figsize=(12,8))
    
    p1, = axs.plot(domain, dens_apo(domain), alpha=0.8, linewidth=4, label='Apo')
    axs.fill_between(domain, dens_apo(domain), alpha=0.2)
    
    p2, = axs.plot(domain, dens_nat(domain), alpha=0.8, linewidth=4, label='Peptide')
    axs.fill_between(domain, dens_nat(domain), alpha=0.2)
    
    p3, = axs.plot(domain, dens_nirm(domain), alpha=0.8, linewidth=4, label='Covalent')
    axs.fill_between(domain, dens_nirm(domain), alpha=0.2)
    
    p4, = axs.plot(domain, dens_ens(domain), alpha=0.8, linewidth=4, label='Noncovalent')
    axs.fill_between(domain, dens_ens(domain), alpha=0.2)
    
    plt.ylabel("Density")
    plt.xlabel("Interface Pocket Volume (^3)")
    plt.title(title)
    plt.legend(handles=[p1, p2, p3, p4])
    fig.tight_layout()
    
```

```python
def plot_distribution_bysystem(apo_dimer,bound_dimer,bound_asym,title):
    params = {
        'legend.fontsize': '12',
        'axes.labelsize': '18',
         'axes.titlesize': '24',
         'xtick.labelsize': '18',
         'ytick.labelsize': '18'
    }

    pylab.rcParams.update(params)
    
    #Set range
    collection = apo_dimer + bound_dimer + bound_asym
    start = np.min(collection) - 1
    stop = np.max(collection) + 1
    
    domain = np.linspace(start,stop,10000)
    
    #Calculate distribution
    dens_apo = gaussian_kde(apo_dimer)
    dens_apo.covariance_factor = lambda : 0.25
    
    dens_dimer = gaussian_kde(bound_dimer)
    dens_dimer.covariance_factor = lambda : 0.25
    
    dens_asym = gaussian_kde(bound_asym)
    dens_asym.covariance_factor = lambda : 0.25
    
    #Plot
    fig, axs = plt.subplots(1,1, sharex=True, sharey=True, tight_layout=True, figsize=(12,8))
    
    p1, = axs.plot(domain, dens_apo(domain), alpha=0.8, linewidth=4, label='Apo dimer', color='darkblue')
    axs.fill_between(domain, dens_apo(domain), alpha=0.2,color='darkblue')
    
    p2, = axs.plot(domain, dens_dimer(domain), alpha=0.8, linewidth=4, label='Full dimer', color='gray')
    axs.fill_between(domain, dens_dimer(domain), alpha=0.2,color='gray')
    
    p3, = axs.plot(domain, dens_asym(domain), alpha=0.8, linewidth=4, label='Half Dimer', color='silver')
    axs.fill_between(domain, dens_asym(domain), alpha=0.2,color='silver')
    
    plt.ylabel("Density")
    plt.xlabel("Interface Pocket Volume (^3)")
    plt.title(title)
    plt.legend(handles=[p1, p2, p3])
    fig.tight_layout()
```

## Nat dimer/asym interface

```python
plot_distribution_bysystem(apo_dimer_int_10A, nat_dimer_int_10A, nat_asym_int_10A,'Mpro + Natural Substrate')
```
