```python
def plot_rmsf_wt_dm(wt_data,dm_data,title):
    params = {
         'font.sans-serif': 'Arial',
         'legend.fontsize': '20',
         'axes.labelsize': '36',
         'axes.titlesize': '36',
         'xtick.labelsize': '20',
         'ytick.labelsize': '20'
    }
    pylab.rcParams.update(params)
    
    fig = plt.figure(figsize=(10,8))
    
    
    ## take care of wt data first
    # average chain A and B together
    wt_chainAB = []
    for i,j in zip(wt_data[1][0:301],wt_data[1][306:]):
        wt_chainAB.append(np.mean((i,j)))

    ## take care of dm data next
    # first have to renumber to match wt
    wtnum_dmdata=[]
    for i in range(len(dm_data[1])):
        if i == 167:
            wtnum_dmdata.append(np.mean((dm_data[1][166],dm_data[1][167])))
            wtnum_dmdata.append(dm_data[1][167])
        elif i == 472:
            wtnum_dmdata.append(np.mean((dm_data[1][471],dm_data[1][472])))
            wtnum_dmdata.append(dm_data[1][472])
        else:
            wtnum_dmdata.append(dm_data[1][i])
    wt_resnums=range(1,608)
    # now average chain A and B together
    dm_chainAB = []
    for i,j in zip(wtnum_dmdata[0:301],wtnum_dmdata[306:]):
        dm_chainAB.append(np.mean((i,j)))
    
    plt.plot(wt_data[0][306:], wt_chainAB,lw=5,color='k',label="WT")
    plt.plot(wt_resnums[0:301], dm_chainAB,lw=5,color='g',label="DM")
    
    plt.ylabel("RMSF ($\AA{}$)")
    plt.xlabel("Residue Number")
    plt.title(title)
    #plt.xlim([0,170])
    plt.legend()
    plt.savefig("{}_rmsf_chainAB.png".format(title),dpi=300)
```

```python
def calc_rmsf(u_traj,u_ref):
    cas=u_traj.select_atoms("protein and name CA")
    prealigner = align.AlignTraj(u_traj, u_ref, select="protein and name CA", in_memory=True).run()
    reference_coordinates=u_traj.trajectory.timeseries(asel=cas).mean(axis=1)
    reference=mda.Merge(cas).load_new(reference_coordinates[:,None,:], order="afc")
    aligner=align.AlignTraj(u_traj,reference,select="protein and name CA",in_memory=True).run()
    rmsfer = RMSF(cas,verbose=True).run()
    return(cas.resnums, rmsfer.rmsf)
```

```python
def calc_avg_rmsfs(all_data):
    resnums=np.array(range(1,302))
    avg=[]
    stdev=[]
    rep1_chainA,rep2_chainA,rep3_chainA,rep1_chainB,rep2_chainB,rep3_chainB = all_data
    for i,j,k,l,m,n in zip(rep1_chainA[1][0:301],rep2_chainA[1][0:301],rep3_chainA[1][0:301],rep1_chainB[1][0:301],rep2_chainB[1][0:301],rep3_chainB[1][0:301]):
        avg.append(np.mean((i,j,k,l,m,n)))
        stdev.append(np.std((i,j,k,l,m,n)))
    avg=np.array(avg)
    stdev=np.array(stdev)
    return(resnums,avg,stdev)
```

```python
def plot_avg_rmsfs(data,title):
    params = {
         'font.family': 'sans-serif',
         'font.sans-serif': 'Arial',
         'legend.fontsize': '20',
         'axes.labelsize': '36',
         'axes.titlesize': '36',
         'xtick.labelsize': '20',
         'ytick.labelsize': '20'
    }
    pylab.rcParams.update(params)
    
    fig = plt.figure(figsize=(10,8))
    resnums,avg,stdev=data
    plt.plot(resnums,avg,lw=5,color='b')
    plt.fill_between(resnums,avg-stdev,avg+stdev,color='b',alpha=0.2)

    plt.ylabel("RMSF ($\AA{}$)")
    plt.xlabel("Residue Number")
    plt.title(title)
    #plt.xlim([0,170])
    #plt.legend()
    plt.savefig("{}_rmsf.png".format(title),dpi=300)
```

```python
def plot_avg_rmsfs(wt_data,dm_data,delP168_data,a173v_data,title):
    params = {
         'font.family': 'sans-serif',
         'font.sans-serif': 'Arial',
         'legend.fontsize': '20',
         'axes.labelsize': '36',
         'axes.titlesize': '36',
         'xtick.labelsize': '20',
         'ytick.labelsize': '20'
    }
    pylab.rcParams.update(params)
    
    wt_resnums,wt_avg,wt_stdev=wt_data
    dm_resnums,dm_avg,dm_stdev=dm_data
    delP168_resnums,delP168_avg,delP168_stdev=delP168_data
    a173v_resnums,a173v_avg,a173v_stdev=a173v_data

    ## take care of dm data next
    # first have to renumber to match wt
    dm_avg2=[]
    dm_stdev2=[]
    for i in range(len(dm_avg)):
        if i == 167:
            dm_avg2.append(np.mean((dm_avg[166],dm_avg[167])))
            dm_avg2.append(dm_avg[167])
            dm_stdev2.append((np.sqrt((dm_stdev[166]/2)**2+(dm_stdev[167]/2)**2)))
            dm_stdev2.append(dm_stdev[167])
        else:
            dm_avg2.append(dm_avg[i])
            dm_stdev2.append(dm_stdev[i])
    dm_resnums2=range(1,302)
    dm_avg2=np.array(dm_avg2)
    dm_stdev2=np.array(dm_stdev2)
    
    ## take care of delP168 data next
    # first have to renumber to match wt
    delP168_avg2=[]
    delP168_stdev2=[]
    for i in range(len(delP168_avg)):
        if i == 167:
            delP168_avg2.append(np.mean((delP168_avg[166],delP168_avg[167])))
            delP168_avg2.append(delP168_avg[167])
            delP168_stdev2.append((np.sqrt((delP168_stdev[166]/2)**2+(delP168_stdev[167]/2)**2)))
            delP168_stdev2.append(delP168_stdev[167])
        else:
            delP168_avg2.append(delP168_avg[i])
            delP168_stdev2.append(delP168_stdev[i])
    delP168_resnums2=range(1,302)
    delP168_avg2=np.array(delP168_avg2)
    delP168_stdev2=np.array(delP168_stdev2)
    
    fig, ax = plt.subplots(1,1, sharex=True, sharey=True, tight_layout=True, figsize=(8,8))
    
    p1, = ax.plot(wt_resnums,wt_avg,lw=4,color='cornflowerblue',label='WT')
    ax.fill_between(wt_resnums,wt_avg-wt_stdev,wt_avg+wt_stdev,color='cornflowerblue',alpha=0.1)
    p2, = ax.plot(dm_resnums2,dm_avg2,lw=4,color='burlywood',label='DM')
    ax.fill_between(dm_resnums2,dm_avg2-dm_stdev2,dm_avg2+dm_stdev2,color='burlywood',alpha=0.1)
    p3, = ax.plot(delP168_resnums2,delP168_avg2,lw=4,color='peru',label="$\Delta$P168")
    ax.fill_between(delP168_resnums2,delP168_avg2-delP168_stdev2,delP168_avg2+delP168_stdev2,color='peru',alpha=0.1)
    p4, = ax.plot(a173v_resnums,a173v_avg,lw=4,color='saddlebrown',label='A173V')
    ax.fill_between(a173v_resnums,a173v_avg-a173v_stdev,a173v_avg+a173v_stdev,color='saddlebrown',alpha=0.1)
    ax.set_ylabel("RMSF ($\AA{}$)")
    ax.set_xlabel("Residue Number")

    left,bottom,width,height = [0.4,0.65,0.3,0.3]
    ax1 = fig.add_axes([left,bottom,width,height])
    ax1.plot(wt_resnums,wt_avg,lw=4,color='cornflowerblue',label='WT')
    ax1.fill_between(wt_resnums,wt_avg-wt_stdev,wt_avg+wt_stdev,color='cornflowerblue',alpha=0.1)
    ax1.plot(dm_resnums2,dm_avg2,lw=4,color='burlywood')
    ax1.fill_between(dm_resnums2,dm_avg2-dm_stdev2,dm_avg2+dm_stdev2,color='burlywood',alpha=0.1)
    ax1.plot(delP168_resnums2,delP168_avg2,lw=4,color='peru')
    ax1.fill_between(delP168_resnums2,delP168_avg2-delP168_stdev2,delP168_avg2+delP168_stdev2,color='peru',alpha=0.1)
    ax1.plot(a173v_resnums,a173v_avg,lw=4,color='saddlebrown')
    ax1.fill_between(a173v_resnums,a173v_avg-a173v_stdev,a173v_avg+a173v_stdev,color='saddlebrown',alpha=0.1)
    ax1.set_xlim([40,70])
    ax1.set_ylim([0,6])
    
    left,bottom,width,height = [0.75,0.65,0.2,0.3]
    ax2 = fig.add_axes([left,bottom,width,height])
    ax2.plot(wt_resnums,wt_avg,lw=4,color='cornflowerblue')
    ax2.fill_between(wt_resnums,wt_avg-wt_stdev,wt_avg+wt_stdev,color='cornflowerblue',alpha=0.1)
    ax2.plot(dm_resnums2,dm_avg2,lw=4,color='burlywood')
    ax2.fill_between(dm_resnums2,dm_avg2-dm_stdev2,dm_avg2+dm_stdev2,color='burlywood',alpha=0.1)
    ax2.plot(delP168_resnums2,delP168_avg2,lw=4,color='peru')
    ax2.fill_between(delP168_resnums2,delP168_avg2-delP168_stdev2,delP168_avg2+delP168_stdev2,color='peru',alpha=0.1)
    ax2.plot(a173v_resnums,a173v_avg,lw=4,color='saddlebrown')
    ax2.fill_between(a173v_resnums,a173v_avg-a173v_stdev,a173v_avg+a173v_stdev,color='saddlebrown',alpha=0.1)
    ax2.set_xlim([180,200])
    ax2.set_ylim([0,6])
    
    fig.tight_layout()
    lgd=plt.legend(handles=[p1, p3, p4, p2], bbox_to_anchor=(1.25,-2.1),ncol=4,fancybox=True)
    plt.savefig("{}_rmsf.png".format(title),bbox_extra_artists=(lgd,), bbox_inches='tight',dpi=300)
```
