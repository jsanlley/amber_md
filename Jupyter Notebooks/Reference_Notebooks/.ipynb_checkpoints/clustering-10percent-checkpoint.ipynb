```python
def calc_pocket(u):
    coors=[]
    pocket=u.select_atoms("protein and resid 24-28 39-50 116-119 141-147 167 171-175 186-192 and name CA")
    natoms=len(pocket)
    for ts in u.trajectory:
        coors.append(pocket.positions)
    #coors=np.array(coors).reshape(natoms)
    return(coors)
```

```python
def calc_pocket_rmsd(u_traj,u_ref):
    R = mda.analysis.rms.RMSD(u_traj,u_ref,
                             select="protein and resid 1-167 169-300 and name CA",
                             groupselections=["protein and resid 24-28 39-50 116-119 141-147 167 171-175 186-192 and name CA"])
    R.run(start=0,stop=330000,step=100)
    return(R.rmsd.T)
```

```python
def cluster_kmeans(data,title):
    data=np.array(data).reshape(-1,1)
    sse = []
    silhouette_coefficients = []
    kmeans_kwargs = {"init": "random","n_init": 10,"max_iter": 300,"random_state": 42,}
    for k in range(1, 100):
        kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
        kmeans.fit(data)
        sse.append(kmeans.inertia_)
        if k > 1:
            score = silhouette_score(data, kmeans.labels_)
            silhouette_coefficients.append(score)
    plot_sse(sse,title)
    plot_silhouette(silhouette_coefficients,title)
    kl = KneeLocator(range(1, 100), sse, curve="convex", direction="decreasing")
    nclusters_elbow=kl.elbow
    nclusters_silhouette=silhouette_coefficients.index(max(silhouette_coefficients))+2
    kmeans_clusters_elbow = KMeans(init="random",n_clusters=nclusters_elbow,n_init=10,max_iter=300,random_state=42)
    kmeans_clusters_elbow.fit(data)
    kmeans_clusters_silhouette = KMeans(init="random",n_clusters=nclusters_silhouette,n_init=10,max_iter=300,random_state=42)
    kmeans_clusters_silhouette.fit(data)
    print("Number of clusters from elbow = {}".format(nclusters_elbow))
    print("Number of clusters from max silhouette coefficient = {}".format(nclusters_silhouette))
    print("Returning as pair: kmeans collected from {} clusters, kmeans collected from {} clusters)".format(nclusters_elbow,nclusters_silhouette))
    return(kmeans_clusters_elbow,kmeans_clusters_silhouette)
```

```python
def plot_sse(sse,title):
    params = {
          'font.sans-serif': 'Arial',
         'legend.fontsize': '20',
         'axes.labelsize': '36',
         'axes.titlesize': '36',
         'xtick.labelsize': '20',
         'ytick.labelsize': '20'
    }
    pylab.rcParams.update(params)
    
    fig = plt.figure(figsize=(12,8))
    
    plt.plot(range(1, 100), sse, linestyle='--',linewidth='4',color='grey')
    plt.xticks(range(1, 100, 10))
    kl = KneeLocator(range(1, 100), sse, curve="convex", direction="decreasing")
    plt.plot(kl.elbow,sse[kl.elbow-1],marker='o',markersize='15',color='black')
    plt.xlabel("Number of Clusters")
    plt.ylabel("SSE ($\AA$)")
    plt.xlim([0,100])
    plt.title("Point of inflection: $k$ = {}".format(kl.elbow))
    #plt.annotate("Point of inflection: $k$ = {}".format(kl.elbow),xy=(kl.elbow,2000000),xycoords='data',xytext=(kl.elbow,5000000),arrowprops=dict(facecolor='black', shrink=0.05),fontsize=20,annotation_clip=False,color='black',bbox=dict(facecolor='white', alpha=0.8,edgecolor='black', boxstyle='round,pad=0.3'))
    fig.tight_layout()
    plt.savefig('{}_sse.png'.format(title),dpi=300) #example
```

```python
def plot_silhouette(silhouette_coefficients,title):
    params = {
         'font.sans-serif': 'Arial',
         'legend.fontsize': '20',
         'axes.labelsize': '36',
         'axes.titlesize': '36',
         'xtick.labelsize': '20',
         'ytick.labelsize': '20'
    }
    pylab.rcParams.update(params)
    
    fig = plt.figure(figsize=(10,8))
    
    plt.plot(range(2, 100), silhouette_coefficients, linestyle='--',linewidth='4',color='grey')
    plt.xticks(range(2, 100, 10))
    plt.plot(silhouette_coefficients.index(max(silhouette_coefficients))+2,max(silhouette_coefficients),marker='o',markersize='15',color='black')
    plt.xlim([0,100])
    plt.title("Most distinguished clusters: $k$ = {}".format(silhouette_coefficients.index(max(silhouette_coefficients))+2)) 
    plt.xlabel("Number of Clusters")
    plt.ylabel("Silhouette Coefficient")
    fig.tight_layout()
    plt.savefig('{}_silhouette_coefs.png'.format(title),dpi=300) #example
```

```python
def plot_counts_per_cluster(counts,title,color):
    params = {
        #'font.family': 'sans-serif',
        #'font.sans-serif': 'Arial',
         'legend.fontsize': '36',
         #'figure.figsize': (24,12),
         'axes.labelsize': '40',
         'axes.titlesize': '40',
         'xtick.labelsize': '24',
         'ytick.labelsize': '24'
         #'xtick.size': '30'
         #'ytick.size': '30'
    }
    
    pylab.rcParams.update(params)
    fig=plt.figure(figsize=(10,10))
    X=range(0,len(counts))
    
    plt.bar(X, counts, color=color, width=0.6)
    plt.xlabel("Cluster Number")
    plt.xticks(X)
    plt.ylabel("Cluster Population")
    plt.tight_layout()
    plt.title("#Clusters and Population for {}".format(title))
    plt.savefig("{}_cluster_population.png".format(title),dpi=300)
```

```python
def plot_cluster_over_time(clusters,title,color):
    params = {
        #'font.family': 'sans-serif',
        #'font.sans-serif': 'Arial',
         'legend.fontsize': '36',
         #'figure.figsize': (24,12),
         'axes.labelsize': '40',
         'axes.titlesize': '40',
         'xtick.labelsize': '24',
         'ytick.labelsize': '24'
         #'xtick.size': '30'
         #'ytick.size': '30'
    }
    
    pylab.rcParams.update(params)
    fig=plt.figure(figsize=(10,10))
    X=range(0,len(clusters.labels_))
    
    plt.plot(X,clusters.labels_, color=color, linewidth=3)
    plt.xlabel("Step Number")
    plt.ylabel("Cluster Population")
    plt.tight_layout()
    plt.title("#Clusters over Time for {}".format(title))
    plt.savefig("{}_clusters_vs_time.png".format(title),dpi=300)
```

```python
def find_cluster_frames(data,clusters,title,color):
    nclusters=len(clusters.cluster_centers_)
    print("working on {}".format(title))
    total_count=[]
    for i in range(nclusters):
        count=sum(1 for l in clusters.labels_ if l == i)
        total_count.append(count)
    print("Counts per cluster:")
    print(total_count)
    plot_cluster_over_time(clusters,title,color)
    plot_counts_per_cluster(total_count,title,color)
    #cluster1=kmeans.cluster_centers_[1]
    cluster_centers=[]
    for i in range(nclusters):
        cluster_centers.append(clusters.cluster_centers_[i]) 
    frames=[]
    for i in range(nclusters):
        print("Cluster {}".format(i))
        print("Center is: {}".format(cluster_centers[i]))
        #frame=data.index(cluster_centers[i])
        frame_center=data.index(get_dist_to_center(data,cluster_centers[i]))
        if frame_center >= 3300:
            chain="B"
            frame=frame_center-3300
            frames.append(frame_center)
            print("Frame found! Chain is {}, Frame is {}".format(chain,frame))
        else:
            chain="A"
            frame=frame_center
            frames.append(frame_center)
            print("Frame found! Chain is {}, Frame is {}".format(chain,frame))
```
